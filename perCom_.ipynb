**Kombinasyon**

11 Malzeme bulunan bir mutfakta 4 malzemeli bir yemek hazırlanacaktır kaç farklı yemek hazırlanabilir ?



```python
from itertools import combinations as cb

def farkli_yemek_sayisi(malzemeSayisi, yemekSayisi):
    # Malzemeler arasından belirli sayıda malzeme seçerek yemek hazırlama kombinasyonunu hesapla
    cbs = list(cb(range(malzemeSayisi), yemekSayisi))
    return len(cbs)
```

```python
depo = ['Domates','Biber','Patlıcan','Soğan','Pırasa','Fasulye','Pırasa','Lahana','Bamya','Kereviz','Brokoli']
yemekMalzemeSayi = 4

yemekList = cb(depo, yemekMalzemeSayi)

yemekSayisi = farkli_yemek_sayisi(len(depo), yemekMalzemeSayi)
print(f"{yemekSayisi} farklı yemek hazırlanabilir.")

print('Yemek listesi : \n')
for x in yemekList:
    print(x)
```

**Kombinasyon 2**

Aynı mutfakta bir menü hazırlanmak istiyor menüde bir çorba bir et yemeği bir sebze yemeği ve iki farklı yan ürün olması gerektiğine göre kaç farklı menü oluşturulabilir ?

*   Çorbalar = Mercimek , Domates , Ezogelin , Sebze , Yayla
*   Et yemekleri = Köfte , Tas Kebabı , Et haşlama , Tavuk Kapama , Mevsim Balığı ,
*   Sebze yemekleri = Portakallı kereviz , Ispanak , Kabak Haşlama , Mücver , Falafel
*   Yan Ürünler = Pilav , Makarna , Ayran , Yoğurt , Meyve , Tatlı





1.   5x5x5xC(6,2) (Çorba * Et * Sebze * İki farklı yan ürün)




```python
from math import comb

corbalar = ["Mercimek", "Domates", "Ezogelin", "Sebze", "Yayla"]
et_yemekleri = ["Köfte", "Tas Kebabı", "Et haşlama", "Tavuk Kapama", "Mevsim Balığı"]
sebze_yemekleri = ["Portakallı kereviz", "Ispanak", "Kabak Haşlama", "Mücver", "Falafel"]
yan_urunler = ["Pilav", "Makarna", "Ayran", "Yoğurt", "Meyve", "Tatlı"]

farkli_menu_sayisi = comb(len(corbalar), 1) * comb(len(et_yemekleri), 1) * comb(len(sebze_yemekleri), 1) * comb(len(yan_urunler), 2)

print(f"{farkli_menu_sayisi} farklı menü oluşturulabilir.")
```

**Permütasyon**

4 kişinin yarıştığı bir koşu yarışmasında sıralı olarak sonuçlar kaç farklı şekilde olabilir ?

```python
from itertools import permutations

# Harfler
yarismacilar = ['Mehmet', 'Ali', 'Zeki', 'Fuat']

# Harflerin permütasyonlarını hesapla
permütasyonlar = permutations(yarismacilar)

# Permütasyonları yazdır
print("Eldeki harflerin permütasyonları:")
for permütasyon in permütasyonlar:
    print(permütasyon)

# Permütasyon sayısını bul
permütasyonlar = permutations(yarismacilar)
permütasyon_sayısı = len(list(permütasyonlar))

print(f"\nToplamda {permütasyon_sayısı} farklı sonuç vardır.")
```
